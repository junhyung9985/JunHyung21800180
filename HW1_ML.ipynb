```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
import mnist_data_python3
from numpy.linalg import eig
```

```python
train_set, val_set, test_set = mnist_data_python3.load_data('mnist.pkl.gz')

train_x, train_y = train_set
val_x, val_y = val_set
test_x, test_y = test_set

print(train_x.shape)
print(train_y.shape)
```

```python
mean_img = train_x.mean(0).reshape(28,28)
```

```python
var_img = train_x.var(0).reshape(28,28)
```

```python
plt.imshow(mean_img, 'gray')
plt.show()
```

```python
plt.imshow(var_img, 'gray')
plt.show()
```

```python
cov = np.cov(train_x.T)

eigval, eigvec = eig(cov)
print('Eigenvector types and eigenvalue before: {}, {}'.format(eigval.dtype, eigvec.dtype))
eigval, eigvec = eigval.astype(np.float32), eigvec.astype(np.float32)
print('Eigenvector types and eigenvalue after: {}, {}'.format(eigval.dtype, eigvec.dtype))
```

```python
plt.plot(range(len(eigval[:100])), eigval[:100])
plt.show()
```

```python
print('Shape of eigenvector before reshape: {}'.format(eigvec.shape))
eigvec3d = eigvec.reshape(28,28,-1)
print('Share of eigenvector after reshape: {}'.format(eigvec3d.shape))
```

```python
fig = plt.figure(figsize = (5,10))
for i in range(10):
    fig.add_subplot(5, 2, i + 1)
    plt.imshow(eigvec3d[:, :, i], 'gray')
plt.show()
```
